```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) # Filled missing values by categorical median value
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# Col = Age Group Column
x = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = x)
```

```python
df[['age_group','age']].head(10)
```

```python
# Col = Purchase Frequency Dayas
frequency= {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
